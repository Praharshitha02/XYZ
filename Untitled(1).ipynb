```python
url = "https://raw.githubusercontent.com/Praharshitha02/XYZ/main/sample1.csv"
names = ['AccX', 'AccY', 'AccZ', 'AccRad', 'SoundP1', 'SoundP2','State']
dataframe = pd.read_csv(url, names=names)
array = dataframe.values
X = array[:,0:6]
Y = array[:,6]
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
clf = GridSearchCV(DecisionTreeClassifier(), cv=5, scoring='roc_auc')
kfold = model_selection.KFold(n_splits=10)
cart = DecisionTreeClassifier()
num_trees = 100
model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
x_actual, x_pred = Y, clf.predict(X)
precision_score_DT_train = precision_score(x_actual, x_pred)
accuracy_score_DT_train = accuracy_score(x_actual, x_pred)
print('The precision score of decision tree on TRAIN is : ',round(precision_score_DT_train * 100,2), '%')
print('The accuracy score of decision tree on TRAIN is : ',round(accuracy_score_DT_train * 100,2), '%')
```

```python
from sklearn.ensemble import RandomForestClassifier
num_trees = 100
max_features = 6
kfold = model_selection.KFold(n_splits=10)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```
